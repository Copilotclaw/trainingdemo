# 🚀 ragmcp – Getting Started

This notebook is your entry point into the [schuerstedt/ragmcp](https://github.com/schuerstedt/ragmcp) project — a RAG (Retrieval-Augmented Generation) system powered by MCP (Model Context Protocol).

## Why clone the full repo?

The notebooks in ragmcp depend on shared Python files (`server.py`, `tool_loader.py`, worker scripts, etc.).
Opening a single notebook without those files will cause import errors.

**Solution:** Clone the whole repo at the start of your session, then open any notebook from there.

## Step 1 – Clone the repository

Run the cell below. It clones the full `ragmcp` repo into the current Colab session (or local environment).
After cloning, all shared files are available.

````python
import os

REPO_URL = "https://github.com/schuerstedt/ragmcp"
REPO_DIR = "ragmcp"

if not os.path.exists(REPO_DIR):
    !git clone {REPO_URL}
    print("✅ Cloned successfully!")
else:
    print("✅ Repo already present, pulling latest changes...")
    !git -C {REPO_DIR} pull

%cd {REPO_DIR}
print(f"\n📂 Working directory: {os.getcwd()}")
print("\nFiles:")
!ls -1
````

## 💡 How Google Colab works with GitHub repos

**Important to know:**

- Colab opens **one notebook at a time** — you can't open a whole folder as a project
- Each Colab session is a **fresh VM** — files from previous sessions are gone
- That's why every notebook should start with a `git clone` to get the shared helper files

**Pattern to add at the top of any ragmcp notebook:**

```python
import os
if not os.path.exists('ragmcp'):
    !git clone https://github.com/schuerstedt/ragmcp
%cd ragmcp
```

This is idempotent — safe to re-run, won't clone twice.
