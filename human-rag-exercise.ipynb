## 🤖 Step 5 — Meet Your Digital Colleague: RAG

What you just did manually is **exactly** what a RAG system does — automatically, in milliseconds, at any scale.

**RAG = Retrieval-Augmented Generation**

Let's map your paper job to the system:

| Your paper job | RAG system component |
|---|---|
| The 10,000-page binder | **Document Store** (the corpus) |
| Thumbing through to find relevant pages | **Retrieval** (similarity search) |
| Deciding "this article is close enough" | **Embedding similarity score** |
| Your mental index / sticky tabs | **Vector Index** (FAISS, Pinecone, etc.) |
| Writing the reply using only the found pages | **Generation with grounded context** |
| Resisting the urge to guess (Enquiry 4) | **Hallucination prevention / grounding** |
| "Not found, I can't answer" | **Fallback / low-confidence response** |

---

### The Problem You Experienced at Scale

With 20 articles, scanning was manageable. But imagine 10,000 pages:

- You can't read them all for every query → you need an **index**
- Your mental index uses keywords → but keyword matching misses synonyms, context, meaning
- A vector embedding captures **meaning**, not just keywords → that's why RAG uses ML embeddings

```
Customer query: "my dog won't stop barking when I leave"
                              ↓
                   embed into vector space
                              ↓
       search 10,000 vectors for nearest neighbours
                              ↓
    returns: [DOG-BEH-02: Separation Anxiety] ← "excessive barking" = match!
                              ↓
       LLM writes reply using ONLY that context
```

No keyword match on "barking" was needed — the **meaning** of the query matched the meaning of the article.

## ⚡ Step 6 — Now Let the Machine Do It

Let's build the automated version of your paper job — a RAG system over the same 20 articles.

````python
# Install dependencies (Colab / fresh environment)
# !pip install sentence-transformers faiss-cpu -q
````

````python
import numpy as np
from sentence_transformers import SentenceTransformer
import faiss

# Load embedding model (same as used in simple-rag-demo)
print("Loading embedding model...")
model = SentenceTransformer("all-MiniLM-L6-v2")

# Embed all 20 articles
texts = [f"{a['title']}. {a['text']}" for a in knowledge_base]
embeddings = model.encode(texts, show_progress_bar=True)

# Build FAISS index
dim = embeddings.shape[1]
index = faiss.IndexFlatL2(dim)
index.add(np.array(embeddings).astype("float32"))

print(f"\n✅ Indexed {index.ntotal} articles. Ready for queries!")
````

````python
def retrieve(query: str, k: int = 3) -> list:
    """Find the top-k most relevant articles for a query."""
    q_vec = model.encode([query]).astype("float32")
    distances, indices = index.search(q_vec, k)
    results = []
    for dist, idx in zip(distances[0], indices[0]):
        article = knowledge_base[idx]
        results.append({
            "id": article["id"],
            "title": article["title"],
            "text": article["text"],
            "score": float(dist)  # lower = more similar (L2 distance)
        })
    return results


def answer(query: str):
    """Retrieve + display — the 'R' in RAG."""
    print(f"🔍 Query: {query}")
    print("─" * 60)
    results = retrieve(query)
    for i, r in enumerate(results, 1):
        print(f"{i}. [{r['id']}] {r['title']}  (distance: {r['score']:.3f})")
        print(f"   {r['text'][:120]}...")
        print()
    return results
````

````python
# 🧪 Enquiry 1 — the machine's turn
results = answer("My dog keeps chewing furniture whenever I leave the house. Is this normal?")
````

````python
# 🧪 Enquiry 2
results = answer("My elderly cat has been drinking a lot more water lately and seems to be losing weight.")
````

````python
# 🧪 Enquiry 3
results = answer("My friend is allergic to cats but really wants one. Are there any breeds that might work?")
````

````python
# 🧪 Enquiry 4 — the one NOT in the knowledge base
results = answer("What's the best food brand for a diabetic dog?")
print("\n⚠️  Note: The system still retrieves 'something'. It has no concept of 'I don't know'.")
print("    This is the hallucination risk — the 'G' in RAG needs guardrails.")
````
